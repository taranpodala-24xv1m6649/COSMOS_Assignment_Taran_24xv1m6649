# Single sample t-test

```python
import pandas as pd
from scipy.stats import ttest_1samp

# Load the dataset
df = pd.read_excel("height_weight.xlsx")

# Extract the height column
heights = df["Height (cm)"]

heights
```

```python
from scipy.stats import ttest_1samp
import numpy as np

mu = 170  # Hypothesized mean

t_stat, p_value = ttest_1samp(heights, mu)

print("One Sample t-Test Result")
print("-------------------------")
print("Sample Mean:", round(heights.mean(), 2))
print("Hypothesized Mean:", mu)
print("t-Statistic:", round(t_stat, 4))
print("p-Value:", p_value)

if p_value < 0.05:
    print("\nConclusion: Reject the null hypothesis (significant difference).")
else:
    print("\nConclusion: Fail to reject the null hypothesis (no significant difference).")
```

#  Two Sampled t-test

```python
import pandas as pd
from scipy.stats import ttest_ind
import numpy as np

# Load the new dataset
df = pd.read_excel("test_scores.xlsx")
print(df)

```

```python
# Extract columns as separate groups
classA = df["Class A Score"]
classB = df["Class B Score"]

print("Class A Scores:\n", classA.tolist())
print("\nClass B Scores:\n", classB.tolist())
```

```python
from scipy.stats import ttest_ind

# Perform two-sample independent t-test
t_stat, p_value = ttest_ind(classA, classB)

print("Two-Sample t-Test Result")
print("-------------------------")
print("t-Statistic:", round(t_stat, 4))
print("p-Value:", p_value)

# Interpretation (alpha = 0.05)
if p_value < 0.05:
    print("\nConclusion: Reject the null hypothesis (significant difference).")
else:
    print("\nConclusion: Fail to reject the null hypothesis (no significant difference).")
```

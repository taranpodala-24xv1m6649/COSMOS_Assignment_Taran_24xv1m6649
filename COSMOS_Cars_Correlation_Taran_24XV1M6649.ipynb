```python
import pandas as pd

df = pd.read_excel("cars.xlsx")

# Selecting two columns for covariance and correlation
x = df["engine_size"].tolist()
y = df["horsepower"].tolist()

df.head()
```

# Covariance

Covariance brings about the variation across variables.
We use covariance to measure how much two variables change with each other.

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

# Read Images
img = mpimg.imread('cov.png')

# Output Images
plt.imshow(img)
```

```python
def covariance(x, y):
    # Finding the mean of the series x and y
    mean_x = sum(x)/float(len(x))
    mean_y = sum(y)/float(len(y))
    # Subtracting mean from the individual elements
    sub_x = [i - mean_x for i in x]
    sub_y = [i - mean_y for i in y]
    numerator = sum([sub_x[i]*sub_y[i] for i in range(len(sub_x))])
    denominator = len(x)-1
    cov = numerator/denominator
    return cov
cov_func = covariance(x, y)
print("Covariance from the custom function:", cov_func)
```

# Correlation


Correlation reveals the relation between the variables.
We use correlation to determine how strongly linked two variables are to each other.

```python
# Read Images
img = mpimg.imread('corr.png')

# Output Images
plt.imshow(img)
```

```python
def correlation(x, y):
    # Finding the mean of the series x and y
    mean_x = sum(x)/float(len(x))
    mean_y = sum(y)/float(len(y))
    # Subtracting mean from the individual elements
    sub_x = [i-mean_x for i in x]
    sub_y = [i-mean_y for i in y]
    # covariance for x and y
    numerator = sum([sub_x[i]*sub_y[i] for i in range(len(sub_x))])
    # Standard Deviation of x and y
    Var_x = sum([sub_x[i]**2.0 for i in range(len(sub_x))])
    Var_y = sum([sub_y[i]**2.0 for i in range(len(sub_y))])
    # squaring by 0.5 to find the square root
    denominator = (Var_x*Var_y)**0.5 # short but equivalent to (std_deviation_x**0.5) * (std_deviation_y**0.5)
    cor = numerator/denominator
    return cor

cor_func = correlation(x, y)
print("Correlation from the custom function:", cor_func)
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

# Create new custom dataset
np.random.seed(10)
size = np.random.normal(1400, 300, 200)  # house size (sq ft)
price = 50 + 0.08 * size + np.random.normal(0, 20, 200)  # price model with noise

# Create dataframe
data = pd.DataFrame({
    'size': size,
    'price': price
})

data.head()
```

The method of least squares estimates the conditional mean of the response variable across values of the predictor variables, quantile regression estimates the conditional median (or other quantiles) of the response variable.

```python
mod = smf.quantreg('price ~ size', data)
res = mod.fit(q=0.5)   # median regression
print(res.summary())
```

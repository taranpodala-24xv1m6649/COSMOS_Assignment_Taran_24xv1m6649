```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 3, figsize=(16, 5), sharey=True)

df.plot(kind='scatter', x='TV', y='Sales', ax=axs[0], title='TV vs Sales')
df.plot(kind='scatter', x='Radio', y='Sales', ax=axs[1], title='Radio vs Sales')
df.plot(kind='scatter', x='Newspaper', y='Sales', ax=axs[2], title='Newspaper vs Sales')

plt.tight_layout()
plt.show()
```

```python
import statsmodels.api as sm

# Add constant (intercept)
X_const = sm.add_constant(X)

# Fit the model
model = sm.OLS(y, X_const).fit()

# Show summary
print(model.summary())
```

```python
import statsmodels.formula.api as smf

# create model using formula notation
lm = smf.ols(formula='Sales ~ TV + Radio + Newspaper', data=df).fit()

# print coefficients
lm.params
```

```python
X_new = pd.DataFrame({
    'TV': [50, 100, 200],
    'Radio': [20, 30, 40],
    'Newspaper': [10, 15, 20]
})

lm.predict(X_new)
```
